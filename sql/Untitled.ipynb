```python
schema_sql = """
CREATE TABLE dim_fund (
    amfi_code INTEGER PRIMARY KEY,
    scheme_name TEXT,
    fund_house TEXT,
    category TEXT,
    plan TEXT,
    risk_grade TEXT
);
"""

with open("sql/schema.sql", "w") as f:
    f.write(schema_sql)

print("schema.sql created successfully!")
```
